```python
#Generar una matriz de numpy con la información de:
#Días transcurridos desde la primera fecha que aparece en el DataFrame.
#Muertes de hombres para las respectivas fechas de la anterior columna
#Muertes de mujeres para las respectivas fechas de la anterior columna
#Recuperaciones de hombres para las respectivas fechas de la anterior columna
#Recuperaciones de mujeres para las respectivas fechas de la anterior columna

#GENERAR UNA MATRIZ CON LA INFORMACION DE NUMPY
#Sacamos las muertes las muertes y el numero de muertes entre hombres sy mujeres
import pandas as pd
import numpy as np
df= pd.read_csv ("Casos_positivos_de_COVID-19_en_Colombia_20231026.csv", sep=",")

#Mirar los dias transcurridos que han pasado entre la primera fecha y la ultima
#Primero eliminamos las filas donde hayan fehas repetidas nulas

df["Fecha de notificación"] = pd.to_datetime(df["Fecha de notificación"])
df = df.sort_values(by="Fecha de notificación")
#df["Fecha de notificación"]=df["Fecha de notificación"].sort_values()

df_sin_fechas_rep = df["Fecha de notificación"].dropna() #Aqui hacemos un nuevo df con solo la columna de "Fecha de notificacion"
primera_fecha = df_sin_fechas_rep.max() #fecha mas reciente
ultima_fecha = df_sin_fechas_rep.min() #primera fecha
dias_transcurridos = primera_fecha - ultima_fecha
print(f"Los dias trasncurridos en de la primera a la ultima fecha es {dias_transcurridos}")

resultados = []  # Lista para almacenar los resultados por día

fechas_unicas = df["Fecha de notificación"].head(7).unique()

for i in fechas_unicas :
    muertes_h = df[(df["Sexo"]=="M") & (df["Recuperado"]=="Fallecido") & (df["Fecha de notificación"]== i)].shape[0]
    numero_muertes_h = muertes_h
    muertes_f = df[(df["Sexo"]=="F") & (df["Recuperado"]=="Recuperado") & (df["Fecha de notificación"]== i)].shape[0]
    numero_muertes_f = muertes_f
        
    #Ahora sacamos los recuperados entre hombres y mujeres
    recup_h = df[(df["Sexo"]=="M") & (df["Recuperado"]=="Recuperado") & (df["Fecha de notificación"]== i)].shape[0]
    numero_recup_h = recup_h
    recup_f = df[(df["Sexo"]=="F") & (df["Recuperado"]=="Recuperado") & (df["Fecha de notificación"]== i)].shape[0]
    numero_recup_f = recup_f

    resultados.append([numero_muertes_h, numero_muertes_f, numero_recup_h, numero_recup_f])
    #Con estos datos numericos ya pondemos armar una matriz
    
# Convertir la lista de resultados en una matriz
matriz_resultados = np.array(resultados)

# Mostrar la matriz consolidada
display(matriz_resultados)
```

Metodo 2 sin numpy

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Casos_positivos_de_COVID-19_en_Colombia_20231026.csv", sep=",")

df["Fecha de notificación"] = pd.to_datetime(df["Fecha de notificación"])
df = df.sort_values(by="Fecha de notificación")

resultados = []  # Lista para almacenar los resultados por día

fechas_unicas = df["Fecha de notificación"].head(7).unique()

for i in fechas_unicas:
    datos_fecha = df[df["Fecha de notificación"] == i]
    muertes_h = datos_fecha[(datos_fecha["Sexo"] == "M") & (datos_fecha["Recuperado"] == "Fallecido")].shape[0]
    muertes_f = datos_fecha[(datos_fecha["Sexo"] == "F") & (datos_fecha["Recuperado"] == "Fallecido")].shape[0]
    recup_h = datos_fecha[(datos_fecha["Sexo"] == "M") & (datos_fecha["Recuperado"] == "Recuperado")].shape[0]
    recup_f = datos_fecha[(datos_fecha["Sexo"] == "F") & (datos_fecha["Recuperado"] == "Recuperado")].shape[0]

    resultados.append([i, muertes_h, muertes_f, recup_h, recup_f])

# Mostrar los resultados con la fecha y los recuentos
for resultado in resultados:
    print(f"Fecha: {resultado[0]}, Muertes Hombres: {resultado[1]}, Muertes Mujeres: {resultado[2]}, Recuperados Hombres: {resultado[3]}, Recuperados Mujeres: {resultado[4]}")
```
