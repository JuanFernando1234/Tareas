```python
class GasesIdeales:
    def __init__(self, volumen, presion, temperatura):
        self.V = volumen
        self.P = presion
        self.T = temperatura
        self.R = 8.314  # Usar un punto decimal en lugar de coma para el valor de R

    def calcular_presion(self):
        respuesta = (self.R * self.T) / self.V
        return respuesta


volumen = float(input("Ingrese el volumen cm^3: "))
presion = 0
temperatura = float(input("Ingrese la temperatura K: "))


gas = GasesIdeales(volumen, presion, temperatura)


presion_calculada = gas.calcular_presion()
print(f"La presión en Pa calculada es: {presion_calculada:.5f}")
```

```python
#2. Realice un programa que pase de kilogramos a libras, cerciorándose que se ingrese un dato correcto
def kg_a_libras(kilogramos):
    libras = kilogramos * 2.20462
    return libras


try:
    peso_kg = float(input("Ingrese el peso en kilogramos: "))
    
    peso_libras = kg_a_libras(peso_kg)
   
    print(f"{peso_kg} kilogramos son aproximadamente {peso_libras:.2f} libras.")
except ValueError:
    print("Por favor, ingrese un número válido.")
```

Has un codigo para saber el factorial de un numero usando: ciclo for, usando funciones sin usar ciclo for y usando la funcion lamda.

```python
def factorial_con_for(numero):
    resultado = 1
    for i in range(1, numero + 1):
        resultado *= i
    return resultado


numero_usuario = int(input("Ingrese un número para calcular su factorial: "))


factorial_resultado = factorial_con_for(numero_usuario)
print(f"El factorial de {numero_usuario} es: {factorial_resultado}")
```

```python
def factorial_sin_for(numero):
    if numero == 0 or numero == 1:
        return 1
    else:
        return numero * factorial_sin_for(numero - 1)


numero_usuario = int(input("Ingrese un número para calcular su factorial: "))


factorial_resultado = factorial_sin_for(numero_usuario)
print(f"El factorial de {numero_usuario} es: {factorial_resultado}")
```

```python
factorial_lambda = (lambda n: 1 if n == 0 or n == 1 else n * factorial_lambda(n - 1))


numero_usuario = int(input("Ingrese un número para calcular su factorial: "))


factorial_resultado = factorial_lambda(numero_usuario)
print(f"El factorial de {numero_usuario} es: {factorial_resultado}")
```
